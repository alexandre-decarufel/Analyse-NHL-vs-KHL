```python
# Âge et pic de performance des joueurs

# À quel âge les joueurs atteignent-ils leur pic de performance dans la NHL et la KHL ?

#Ajouter une colonne pour différencier les ligues
nhl_combined_withleague = nhl_combined_adjusted.copy()
khl_combined_withleague = khl_combined_adjusted.copy()

nhl_combined_withleague['League'] = 'NHL'
khl_combined_withleague['League'] = 'KHL'

#Verification
#display(nhl_combined_withleague.head())
#display(khl_combined_withleague.head())

#Combiner les deux
combined = pd.concat([nhl_combined_withleague, khl_combined_withleague], ignore_index=True)

#Garder la meilleure saison de chaque joueur
best_seasons = combined.loc[combined.groupby('Name')['Adjusted_P'].idxmax()]

average_age_at_peak = best_seasons.groupby('League')['Age'].mean().reset_index()
average_age_at_peak = round(average_age_at_peak, 1)

#Vérifier le résultat
#print(best_seasons[['Name', 'Adjusted_P', 'League']].head())
print(f"L'age moyen des meilleures saisons est {average_age_at_peak}")

#Calcul de la performance moyenne pour les joueurs de 25 ans
average_performance_25_nhl = best_seasons[(best_seasons['Age'] == 25) & (best_seasons['League'] == 'NHL')]['Adjusted_P'].mean()
average_performance_25_khl = best_seasons[(best_seasons['Age'] == 25) & (best_seasons['League'] == 'KHL')]['Adjusted_P'].mean()
average_performance_25_nhl = round(average_performance_25_nhl, 3)
average_performance_25_khl = round(average_performance_25_khl, 3)


#Calcul de la performance moyenne pour les joueurs de 26 ans
average_performance_26_nhl = best_seasons[(best_seasons['Age'] == 26) & (best_seasons['League'] == 'NHL')]['Adjusted_P'].mean()
average_performance_26_khl = best_seasons[(best_seasons['Age'] == 26) & (best_seasons['League'] == 'KHL')]['Adjusted_P'].mean()
average_performance_26_nhl = round(average_performance_26_nhl, 3)
average_performance_26_khl = round(average_performance_26_khl, 3)

print(f"La performance moyenne pour les joueurs de 25 ans en NHL est de {average_performance_25_nhl}.")
print(f"La performance moyenne pour les joueurs de 26 ans en NHL est de {average_performance_26_nhl}.")
print(f"La performance moyenne pour les joueurs de 25 ans en KHL est de {average_performance_25_khl}.")
print(f"La performance moyenne pour les joueurs de 26 ans en KHL est de {average_performance_26_khl}.")

import matplotlib.pyplot as plt

# Distribution de l'âge au pic de performance pour chaque ligue
plt.hist(best_seasons[best_seasons['League'] == 'NHL']['Age'], bins=range(15, 50), alpha=0.7, label='NHL')
plt.hist(best_seasons[best_seasons['League'] == 'KHL']['Age'], bins=range(15, 50), alpha=0.7, label='KHL')

plt.xlabel('Âge')
plt.ylabel('Nombre de joueurs')
plt.title('Âge au pic de performance (NHL vs KHL)')
plt.legend()
plt.show()


# Y a-t-il une différence notable dans le déclin des performances avec l’âge dans les deux ligues ?


# Calculer la performance moyenne (Adjusted_P) par âge et par ligue
average_performance_by_age = combined.groupby(['League', 'Age'])['Adjusted_P'].mean().reset_index()

# Filtrer pour les joueurs âgés de 25 ans et plus
average_performance_by_age_25_plus = average_performance_by_age[average_performance_by_age['Age'] >= 25]

# Séparer les données par ligue
nhl_data_25_plus = average_performance_by_age_25_plus[average_performance_by_age_25_plus['League'] == 'NHL']
khl_data_25_plus = average_performance_by_age_25_plus[average_performance_by_age_25_plus['League'] == 'KHL']

# Graphique : Performance moyenne pour les joueurs de 25 ans et plus dans chaque ligue
plt.bar(nhl_data_25_plus['Age'], nhl_data_25_plus['Adjusted_P'], alpha=0.7, label='NHL')
plt.bar(khl_data_25_plus['Age'], khl_data_25_plus['Adjusted_P'], alpha=0.7, label='KHL')

# Ajouter les étiquettes et le titre
plt.xlabel('Âge')
plt.ylabel('Performance moyenne (Adjusted_P)')
plt.title('Performance moyenne par âge (25 ans et plus, NHL vs KHL)')
plt.legend()
plt.show()
```
